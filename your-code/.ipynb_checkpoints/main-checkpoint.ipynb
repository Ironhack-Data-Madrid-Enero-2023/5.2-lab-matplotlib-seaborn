#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or Weekend columns.

```python
mediana = fitbit[['Months_encoded', 'Steps']].groupby('Months_encoded', as_index=False).median()

mediana.info()
```

```python
plt.bar(x=mediana.Months_encoded , height=mediana.Steps)
plt.rcParams['figure.figsize']=(12, 4)
plt.sca
plt.xticks(ticks=mediana['Months_encoded']);
plt.savefig('steps_by_Months_encoded.png')           
```

```python
mediana2= fitbit[['Work_or_Weekend', 'Steps']].groupby('Work_or_Weekend', as_index=False).median()
```

```python
plt.bar(x=mediana2.Work_or_Weekend , height=mediana2.Steps)
plt.rcParams['figure.figsize']=(12, 4)
plt.sca
plt.xticks(ticks=mediana2['Work_or_Weekend']);
plt.savefig('steps_by_week_or_weekend.png')           
```

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
sns.histplot(x=titanic.Age,
             bins=40,
             element='step');   


plt.axvline(titanic.Age.mean(),
            c='r',
            linewidth=3.,
            linestyle='--',
            label='media de edad')

plt.axvline(titanic.Age.median(),
            c='orange',
            linewidth=3.,
            linestyle='--',
            label='mediana de edad')

plt.axvline(titanic.Age.mode()[0],
            c='black',
            linewidth=3.,
            linestyle='--',
            label='moda de edad')

plt.title('Histograma de edades Titanic')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()

plt.show();
```

```python
corr=titanic.corr(method='pearson')
```

#### Plot the correlation matrix using seaborn.

```python
sns.set(style='white')                            

mascara=np.triu(np.ones_like(corr, dtype=bool))   

cmap=sns.diverging_palette(0, 10, as_cmap=True)   


sns.heatmap(corr,            
            mask=mascara,   
            cmap=cmap,      
            vmax=1,          
            vmin=-1,          
            center=0,         
            square=True,      
            linewidth=0.5,    
            
            cbar_kws={'shrink': 0.5},  
            annot=True                 
           );
```
