#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# AVERAGE STEPS BY MONTH_ENCODED
plt.rcParams['figure.figsize']=(12,4) 

f = fitbit.groupby('Months_encoded', as_index = False).median()

plt.plot(f.Months_encoded, f.Steps)
plt.xlabel('MONTHS')
plt.ylabel('STEPS')
plt.xticks(f.Months_encoded)
plt.show()
plt.savefig('../figures/steps_month.png');
```

```python
### AVERAGE STEPS BY WORK_OR_WEEKEND
fitbit.groupby('Work_or_Weekend').median()['Steps'].plot.bar()
plt.ylabel('STEPS')
plt.savefig('../data/steps_work.png');
```

#### Write a loop to plot 3 scatter plots of the following features:

```python
# your code here
for i in ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']:
    fitbit.plot.scatter(x=i, y='Steps')
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
fig, ax = plt.subplots(1, 3, sharey=True) 

titanic.Age.plot.hist(bins = 10, ax=ax[0])

ax[0].set_xlabel('age')

titanic.Age.plot.hist(bins = 20, ax=ax[1])
ax[1].set_xlabel('age')

titanic.Age.plot.hist(bins = 50, ax=ax[2])
ax[2].set_xlabel('age');
fig.supxlabel('AGE'); #Supper X Label
```
