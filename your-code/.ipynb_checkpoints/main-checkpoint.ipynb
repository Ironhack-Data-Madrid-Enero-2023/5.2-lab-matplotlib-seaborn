#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
fitbit.head(5)
```

```python
   # pinta 1 fila 3 columnas, 3 plots en fila
columns = ['Minutes Lightly Active', 'Minutes Fairly Active', 'Minutes Very Active']

for x in columns:
    plt.scatter(fitbit['Steps'], fitbit[x])
    plt.xlabel('Steps')
    plt.ylabel(x)
    plt.title(f'Steps vs {x}')
    plt.show()
```

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
sns.histplot(x=titanic.Age,
             bins=60,
             element='step');   


plt.axvline(titanic.Age.mean(),
            c='blue',
            linewidth=3.,
            linestyle='--',
            label='media de edad')

plt.axvline(titanic.Age.median(),
            c='yellow',
            linewidth=3.,
            linestyle='--',
            label='mediana de edad')

plt.axvline(titanic.Age.mode()[0],
            c='black',
            linewidth=3.,
            linestyle='--',
            label='moda de edad')

plt.title('Histograma de edades Titanic')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()
#plt.savefig('histograma.png')
plt.show();
```

#### Plot the correlation matrix using seaborn.

```python
corr=titanic.corr(method='pearson')
```

```python
sns.set(style='dark')                           

mascara=np.triu(np.ones_like(corr, dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)


sns.heatmap(corr,             
            mask=mascara,    
            cmap=cmap,
            vmax=1,
            vmin=-1,
            center=0,
            square=True,
            linewidth=0.5,
            
            cbar_kws={'shrink': 0.5},
            annot=True
           );
```
