#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks

```python
fit.head()
```

```python
fit['week_end'] = fit['Days'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')
```

```python
median_steps_by_month = fit.groupby('Months_encoded')['Steps'].median()
```

```python
import matplotlib.pyplot as plt
import pandas as pd



# Create a boxplot to visualize the median steps by month
fig, ax = plt.subplots(figsize=(12, 4))
plt.sca(ax)
median_steps_by_month.plot(kind='box')
plt.xticks(range(1, 13), ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
plt.xlabel('Months')
plt.ylabel('Median Steps')
plt.title('Median Steps by Month')
plt.show()

# Calculate the median steps by weekday/weekend
median_steps_by_weekday = fit.groupby('week_end')['Steps'].median()

# Create a bar chart to visualize the median steps by weekday/weekend
fig, ax = plt.subplots(figsize=(12, 4))
plt.sca(ax)
median_steps_by_weekday.plot(kind='bar')
plt.xticks([0, 1], ['Weekend', 'Weekday'])
plt.xlabel('Day Type')
plt.ylabel('Median Steps')
plt.title('Median Steps by Weekday/Weekend')
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

for feature in features:
    plt.scatter(fit[feature], fit['Steps'])
    plt.xlabel(feature)
    plt.ylabel('Steps')
    plt.title(f'{feature} vs Steps')
    plt.show()
```
