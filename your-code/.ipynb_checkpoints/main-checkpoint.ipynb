#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# AVERAGE STEPS BY MONTH_ENCODED
mediana = fitbit[['Months_encoded', 'Steps']].groupby('Months_encoded', as_index=False).median()

plt.figure(figsize=(12, 4))
plt.bar(x=mediana.Months_encoded, height=mediana.Steps)
plt.savefig('../figures/monthsVSsteps.png');
```

```python
# AVERAGE STEPS BY WORK_OR_WEEKEND
mediana = fitbit[['Work_or_Weekend', 'Steps']].groupby('Work_or_Weekend', as_index=False).mean()

plt.figure(figsize=(12, 4))
plt.bar(x=mediana.Work_or_Weekend, height=mediana.Steps)
plt.savefig('../figures/workorweekendVSsteps.png');
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
fig, ax = plt.subplots(3, 1, figsize=(6, 12))

for i,cat in enumerate(fitbit.index[:3]):   
    sns.scatterplot(data=fitbit, x='Minutes Lightly Active', y='Steps', ax=ax[0])

    sns.scatterplot(data=fitbit, x='Minutes Very Active', y='Steps', ax=ax[1])

    sns.scatterplot(data=fitbit, x='Minutes Sedentary', y='Steps', ax=ax[2])
```

#### Plot the correlation matrix using seaborn.

```python
corr = titanic.corr()

f, ax = plt.subplots(figsize=(12, 10))

mask = np.triu(np.ones_like(corr, dtype=bool))

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, annot=True, mask = mask, cmap=cmap);
```
