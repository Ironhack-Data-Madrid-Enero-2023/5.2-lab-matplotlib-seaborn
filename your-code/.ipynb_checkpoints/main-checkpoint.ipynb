```python
mean_steps = fitbit.groupby('Months_encoded')['Steps'].mean()

plt.bar(mean_steps.index, mean_steps.values)
mean_steps.plot.bar(figsize =(12, 4))

plt.show()
```

```python
fitbit['Date'] = pd.to_datetime(fitbit['Date'])
```

```python
# Crear una columna 'Week_or_Weekend' que indique si es un da laborable o de fin de semana
fitbit['Week_or_Weekend'] = fitbit['Date'].apply(lambda x: 'Weekend' if x.weekday()>=5 else 'Weekday')

# Calcular los promedios de pasos por da de la semana
mean_steps1 = fitbit.groupby(['Week_or_Weekend', 'Date'])['Steps'].mean()

# Crear un grfico de barras apilado
mean_steps1.unstack(level=0).plot(kind='bar', stacked=True)
mean_steps1.plot.bar(figsize =(12, 4))

plt.show()
```

```python
plt.savefig('mean_steps1.png', dpi=300)
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
columns = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

for col in columns:
    plt.scatter(fitbit['Steps'], fitbit[col])
    plt.xlabel('Pasos')
    plt.ylabel(col)
    plt.title(f'{col} vs Pasos')
    plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
bins = [10, 20, 50]

for i, bin_size in enumerate(bins):
    axs[i].hist(titanic['Age'], bins=bin_size)
    axs[i].set_xlabel('Age')
    axs[i].set_ylabel('Count')
    axs[i].set_title('Histogram with bin size = {}'.format(bin_size))

plt.show()
```

#### Use the right plot to visualize column `Gender`. There are 2 ways of doing it. Do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
fig, ax = plt.subplots()

# Cuenta la cantidad de veces que aparece cada categora en la columna Gender
counts = titanic['Gender'].value_counts()

# Define una lista de colores para cada gnero
colors = ['blue', 'pink']

# Crea un grfico de barras con la frecuencia de cada categora y colores personalizados
ax.bar(counts.index, counts.values, color=colors)

# Etiqueta los ejes y el ttulo del grfico
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
ax.set_title('Gender Distribution');
```

```python
# Method 2 - seaborn

sns.countplot(data=titanic, x='Gender');
```
