#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4

```python
median_steps_by_month = fitbit.groupby('Months_encoded')['Steps'].median()
median_steps_by_month.reset_index()
```

```python
median_steps_by_month = fitbit.groupby('Work_or_Weekend')['Steps'].median()
median_steps_by_month
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
ax.bar(median_steps_by_month.index, median_steps_by_month.values)
ax.set_xlabel('Months')
ax.set_ylabel('Steps')
ax.set_title('Work_or_Weekend')
```

#### Plot the correlation matrix using seaborn.

```python
corr=titanic.corr(method='pearson')
```

```python
titanic.corr()
```

```python
sns.set(style='white')
mascara=np.triu(np.ones_like(corr, dtype=bool))
cmap=sns.diverging_palette(0, 10, as_cmap=True)

sns.heatmap(corr,           
            mask=mascara,     
            cmap=cmap,        
            vmax=1,           
            vmin=-1,          
            center=0,         
            square=True,      
            linewidth=0.5,    
            
            cbar_kws={'shrink': 0.5},  
            annot=True                 
           );
```

#### What are the most correlated features?

```python
"""
Pclass and Fare

"""
```
