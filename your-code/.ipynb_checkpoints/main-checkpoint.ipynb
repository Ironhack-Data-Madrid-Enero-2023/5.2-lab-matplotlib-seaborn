#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
titanic["Age"].plot(kind = "hist", bins = (len(titanic["Age"].unique())))
plt.axvline(titanic.Age.mean(), color='r')
plt.axvline(titanic.Age.std(), color='g')
plt.axvline(titanic.Age.min(), color='b')
plt.axvline(titanic.Age.quantile(0.25), color='c')
plt.axvline(titanic.Age.quantile(0.55), color='m')
plt.axvline(titanic.Age.quantile(0.75), color='y')
plt.axvline(titanic.Age.max(), color='k');
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
plt.scatter(x=titanic['Age'], y=titanic['Fare']);

```

```python
# Method 2 - seaborn
sns.set_style("whitegrid")
sns.regplot(data=titanic, x="Age", y="Fare")
plt.xlabel('Edad')
plt.ylabel('Tarifa')
plt.show();
```

#### Plot the correlation matrix using seaborn.

```python
corr=titanic.corr(method='pearson')
sns.set(style='white')                            # estilo blanco

mascara=np.triu(np.ones_like(corr, dtype=bool))   # genera una mascara

cmap=sns.diverging_palette(0, 10, as_cmap=True) 
sns.heatmap(corr,             # datos
            mask=mascara,     # mascara
            cmap=cmap,        # color
            vmax=1,           # maximo vertical
            vmin=-1,          # minimo vertical
            center=0,         # centro
            square=True,      # cuadrado el dato pintado
            linewidth=0.5,    # ancho de la linea
            
            cbar_kws={'shrink': 0.5},  # barra de color
            annot=True                 # anotacion, el numero escrito
           );
```
