#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# AVERAGE STEPS BY MONTH_ENCODED
# your code here
asbme = fitbit.groupby('Months_encoded', as_index=False).agg({'Steps':'median'})

asbme
```

```python
plt.bar(x = asbme['Months_encoded'], height = asbme['Steps']);
plt.rcParams['figure.figsize']=(12, 4)
plt.sca
plt.xticks(ticks=asbme['Months_encoded']);
plt.savefig('../figures/Months_encoded.png')
```

```python
# AVERAGE STEPS BY WORK_OR_WEEKEND
# your code here

asbwow = fitbit.groupby('Work_or_Weekend', as_index=False).agg({'Steps':'median'})

asbwow
```

```python
plt.bar(x = asbwow['Work_or_Weekend'], height = asbwow['Steps']);
plt.rcParams['figure.figsize']=(12, 4)
plt.sca
plt.xticks(ticks=asbwow['Work_or_Weekend']);
plt.savefig('../figures/steps_by_week_or_weekend.png')
```

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
# your code here
sns.histplot(x=titanic.Age,
             bins=40,
             element='step');   


plt.axvline(titanic.Age.mean(),
            c='r',
            linewidth=3.,
            linestyle='--',
            label='media')

plt.axvline(titanic.Age.median(),
            c='orange',
            linewidth=3.,
            linestyle='--',
            label='mediana')

plt.axvline(titanic.Age.mode()[0],
            c='black',
            linewidth=3.,
            linestyle='--',
            label='moda')

plt.title('Histograma de edades Titanic')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()
plt.show();
```

#### What additional information does the last plot provide about feature `Age`?

```python
"""
La media est desplazada hacia la derecha con respecto a la moda, es lo que nos muestra la distribucin en la grfica por
efecto de los outliers.
"""
```

```python
corr = titanic.corr(method='pearson')
```

#### Plot the correlation matrix using seaborn.

```python
# your code here
sns.set(style='white')                            # estilo blanco

mascara=np.triu(np.ones_like(corr, dtype=bool))   # genera una mascara

cmap=sns.diverging_palette(0, 10, as_cmap=True)   # paleta de colores


sns.heatmap(corr,             # datos
            mask=mascara,     # mascara
            cmap=cmap,        # color
            vmax=1,           # maximo vertical
            vmin=-1,          # minimo vertical
            center=0,         # centro
            square=True,      # cuadrado el dato pintado
            linewidth=0.5,    # ancho de la linea
            
            cbar_kws={'shrink': 0.5},  # barra de color
            annot=True                 # anotacion, el numero escrito
           );
```

#### What are the most correlated features?

```python
"""
Fare y Pclass de manera inversa.
"""
```
