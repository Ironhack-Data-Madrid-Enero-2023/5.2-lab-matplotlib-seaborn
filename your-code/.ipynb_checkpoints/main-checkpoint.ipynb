#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
sns.histplot(x=titanic.age,
             bins=40,
             element='step');   


plt.axvline(titanic.age.mean(),
            c='lightblue',
            linewidth=3.,
            linestyle='--',
            label='media de edad')

plt.axvline(titanic.age.median(),
            c='lightgreen',
            linewidth=3.,
            linestyle='--',
            label='mediana de edad')

plt.axvline(titanic.age.mode()[0],
            c='pink',
            linewidth=3.,
            linestyle='--',
            label='moda de edad')

plt.title('Histograma de edades Titanic')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()
#plt.savefig('histograma.png')
plt.show();
```

#### What additional information does the last plot provide about feature `Age`?

```python
"""
Here we can more closely appreciate that the majority of passengers were between 20 and 30 years old and there were more passengers older than that rathers than younger than that
"""
```

```python
corr=titanic.corr(method='pearson')
```

#### Plot the correlation matrix using seaborn.

```python
sns.set(style='white')                           

mascara=np.triu(np.ones_like(corr, dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)


sns.heatmap(corr,             
            mask=mascara,    
            cmap=cmap,
            vmax=1,
            vmin=-1,
            center=0,
            square=True,
            linewidth=0.5,
            
            cbar_kws={'shrink': 0.5},
            annot=True
           );
```
